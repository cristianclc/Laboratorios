```python
!pip install basemap

import matplotlib.pyplot as plt
import networkx as nx
import openpyxl
import math
from re import M
from geopy.geocoders import Nominatim
import time
from mpl_toolkits.basemap import Basemap

```

```python
libro = openpyxl.load_workbook(filename="flights_final.xlsx")
hoja = libro.active #primera hoja del excel
```

```python
#latitud norte-sur ----ecuador = 0 ----polo norte = 90 ----polo sur = -90
#longitud oeste-este ----greenwhich = 0 --------oeste = -180  se convierte a --------este = 180
#primer aerpuerto cochin - shajra, distancia 2774 km
nodes = {} #-nombre aeropuerto -[objeto, nombre aeropuerto llegada, distancia]
pos = {} #-nombre aeropuerto --posiciones geográficas
visitados = {} #diccionario de visitados
edges = [] #lista de las aristas

class Node:
    def __init__(self, code, name, ciudad, pais, latitud, longitud):
        self.code = code
        self.name = name
        self.ciudad = ciudad
        self.pais = pais
        self.latitud = latitud
        self.longitud = longitud

class Edge:
    def __init__(self, start, end, weight):
        self.start = start  #aeropuerto de inicio
        self.end = end      #aeropuerto de destino
        self.weight = weight  #peso de la arista (distancia)

class UnionFind: #usado en el kruskal para obtener los componentes de los arboles minimos
    def __init__(self, n):
        self.parent = list(range(n))

    def find(self, node):
        if self.parent[node] != node:
            self.parent[node] = self.find(self.parent[node])
        return self.parent[node]

    def union(self, u, v):
        root_u = self.find(u)
        root_v = self.find(v)

        if root_u != root_v:
            self.parent[root_v] = root_u #los agrega al mismo representante

for fila in hoja.iter_rows(min_row=2, values_only=True):
    # aeropuerto salida
    s_code = str(fila[0])
    s_name = str(fila[1])
    s_ciudad = str(fila[2])
    s_pais = str(fila[3])
    s_latitud = float(fila[4])
    s_longitud = float(fila[5])
    # aeropuerto destino destino
    d_code = str(fila[6])
    d_name = str(fila[7])
    d_ciudad = str(fila[8])
    d_pais = str(fila[9])
    d_latitud = float(fila[10])
    d_longitud = float(fila[11])
    #agregan a la lista de nodos
    node_s = Node(s_code, s_name, s_ciudad, s_pais, s_latitud, s_longitud)
    node_d = Node(d_code, d_name, d_ciudad, d_pais, d_latitud, d_longitud)

    #conversion a radianes para los calculos
    s_latitud = math.radians(s_latitud)
    s_longitud = math.radians(s_longitud)
    d_latitud = math.radians(d_latitud)
    d_longitud = math.radians(d_longitud)
    #calculo distancias resultando en NM multiplicado por 1.852 para convertir en KM
    dis =   int(1.852 * (3440.1 * math.acos((math.sin(s_latitud) * math.sin(d_latitud)) + math.cos(s_latitud) * math.cos(d_latitud) * math.cos(s_longitud - d_longitud))))

    #agregar a diccionario de posiciones
    if s_code not in pos:
        pos[s_code] = (s_longitud, s_latitud)
    if d_code not in pos:
        pos[d_code] = (d_longitud, d_latitud)

    #agregar a diccionario de visitados
    if s_code not in visitados:
        visitados[s_code] = False
    if d_code not in visitados:
        visitados[d_code] = False

                        #objeto, aeropuerto destino, distancia -- agregar a diccionario de conexiones
    if s_code not in nodes: #a aeropuerto le agrega los aerpuertos a los que puede llegar y la distancia
        nodes[s_code] = [node_s, [d_code, dis]]
    else:
        for i in nodes[s_code][1:]:
            if i[0] == d_code:
                break
        else:
            nodes[s_code].append([d_code, dis])
    if d_code not in nodes:
        nodes[d_code] = [node_d, [s_code, dis]]
    else:
        for i in nodes[d_code][1:]:
            if i[0] == s_code:
                break
        else:
            nodes[d_code].append([s_code, dis])

for n in nodes: #crea las aristas
        for i in nodes[n][1:]: #desde el diccionario nodes crea las aristas
            start_node = n
            end_node = i[0]
            distance = i[1]
            edges.append(Edge(start_node, end_node, distance))

def kruskal(nodes, edges):
    mst = []
    uf = UnionFind(len(nodes)) #crea el union find para realizar el algoritmo kruskal y evitar ciclos

    edges.sort(key=lambda edge: edge.weight) #ordena las aristas por peso

    for edge in edges:
        u = nodes.index(edge.start) #número nodo inicio arista del componente
        v = nodes.index(edge.end) #número nodo final arista del componente

        if uf.find(u) != uf.find(v): #busca al representante de cada nodo, si estos son iguales, al unirse daría un ciclo
            uf.union(u, v) #los agrega al componente
            mst.append(edge)

        if len(mst) == len(nodes) - 1: #cuando llega a las n-1 aristas ya esta completo el kruskal
            break

    return mst

def dfs_component(nodo, nodes, visitados, componente): #dfs
    visitados[nodo] = True
    componente.append(nodo)
    for i in nodes[nodo][1:]:
        if not visitados[i[0]]:
            dfs_component(i[0], nodes, visitados, componente)

def encontrar_componentes(nodes, visitados): #lista de 7 componentes
    componentes = [] #lista de componentes
    for nodo in visitados:
        if not visitados[nodo]:
            componente = [] #componente 1 ó 2 ó 3 etc
            dfs_component(nodo, nodes, visitados, componente) #hace el dfs al nodo no visitado y regresa la lista del componente
            componentes.append(componente) #agrega la lista del componente (todos los nodos) a la lista de componentes
    return componentes

def dijkstra(start, target=None): #el target es opcional
    dist = {nodo: float('inf') for nodo in nodes} #distancias del nodo start a todos (empiezan en infinito)
    prev = {node: None for node in nodes} #predecesor de cada nodo para reconstruir el camino

    dist[start] = 0 #distancia a mismo nodo es 0

    unvisited = set(nodes.keys()) #conjunto de nodos no visitados

    while unvisited: #mientras haya nodos no visitados
        current_node = min(unvisited, key=lambda node: dist[node]) #seleccionamos el nodo no visitado con menor peso

        if target and current_node == target: #si llegamos al nodo que buscabamos cierra el ciclo
            break

        unvisited.remove(current_node) #removemos el nodo actual del conjunto de no visitados

        for neighbor, weight in nodes[current_node][1:]: #actualiza las distancias
            if neighbor in unvisited:
                new_dist = dist[current_node] + weight #distancia acumulada del nodo actual mas la distancia al nodo
                if new_dist < dist[neighbor]:
                    dist[neighbor] = new_dist
                    prev[neighbor] = current_node  #predecesor del nodo actual para el camino

    if target: #si hay un nodo destino reconstruye el camino
        path = []
        current = target
        while current is not None:
            path.insert(0, current)  #se ingresa el nodo al inicio del camino
            current = prev[current] #va agregando los predecesores
        return dist, path  #retorna la distancia y el camino hacia el destino específico
    else:
        return dist  #si no hay nodo de llegada, devolvemos solo las distancias con todos los nodos

while True:
  menu = 0
  menu = input("1, para grafo generado es conexo, 2 árbol de expansión mínima, 3 Mostrar información correspondiente del aeropuerto y 10 caminos mas largos, 4 Mostrar  el  camino  mínimo  entre  el  primer  y  el  segundo  vértice, 5 para mostrar el grafo con todos los aeropuertos: ")
  if menu == "1":
    visitados = dict.fromkeys(visitados, False)
    componentes = encontrar_componentes(nodes, visitados) #llama a la funcion para hallar el numero de componentes
    print("Numero de componentes conexas:", len(componentes))
    print("Numero de vertices en cada componente")
    for componente in componentes:
      print(len(componente)) #numero de listas (cada lista es un componente del grafo)
  elif menu == "2":
    visitados = dict.fromkeys(visitados, False) #al diccionario de visitados les pone valores false
    componentes = encontrar_componentes(nodes, visitados)
    print("Kruskal de cada componente")
    for index, componente in enumerate(componentes): #kruskal de cada componente
        subnodos = componente #numero del representante (componente 0, 1, 2 etc)
        subedges = [edge for edge in edges if edge.start in subnodos and edge.end in subnodos] #lista de aristas de la componente actual

        mst = kruskal(subnodos, subedges) #kruskal de cada componente
        sum=0
        for edge in mst:
            sum+=edge.weight
        print(sum) #peso arbol expasion minima de cada componente
  elif menu == "3":
    while True:
      start_vertex = input("Ingrese código del aeropuerto CASE SENSITIVE: ")
      if start_vertex in nodes:
        break
      else:
        print("El aeropuerto no existe")
    print(f"Nombre: {nodes[start_vertex][0].name}, Latitud: {nodes[start_vertex][0].latitud}, Longitud: {nodes[start_vertex][0].longitud}, Ciudad: {nodes[start_vertex][0].ciudad},  Pais: {nodes[start_vertex][0].pais}")
    distances = dijkstra(start_vertex) #djikstra del vertice inicial
    distancias_minimas = sorted([items for items in distances.items() if items[1] < float('inf')], key=lambda x: x[1])[-10:] #ordena las aristas de mayor a menor y se salta las inf
    print("Los 10 aeropuertos con distancias mínimas mas largas son:")
    print(distancias_minimas)
  elif menu ==  "4":
    my_map = Basemap(projection="robin",lon_0=0, lat_0=0)
    while True:
      start_vertex = input("Ingrese código del aeropuerto de salida CASE SENSITIVE: ")
      if start_vertex in nodes:
        break
      else:
        print("El aeropuerto no existe")
    while True:
      target_vertex = input("Ingrese código del aeropuerto de llegada CASE SENSITIVE: ")
      if target_vertex in nodes:
        break
      else:
        print("El aeropuerto no existe")
    distance, path = dijkstra(start_vertex, target_vertex)
    edges_in_path = [(path[i], path[i+1]) for i in range(len(path) - 1)] #lista de aristas del camino mínimo

    if len(path) == 1:
        print("No hay camino posible")
    else:
        for i in path:
            print(f"Nombre: {nodes[i][0].name}, Ciudad: {nodes[i][0].ciudad}, País: {nodes[i][0].pais}, Código: {nodes[i][0].code}, Lat: {nodes[i][0].latitud}, Long: {nodes[i][0].longitud}")
        print(f"Distancia total: {distance[target_vertex]}")

        subgraph = nx.Graph() #va creando el grafo del camino mínimo
        subgraph.add_nodes_from(path)  # añade los nodos del camino
        #pos={}

        for i in range(len(path) - 1): #recorre el path en parejas para agregar las aristias y su peso
            node1 = path[i]
            node2 = path[i+1]

            for target, weight in nodes[node1][1:]: #agarra la distancia desde el diccionario nodes
                if target == node2: #si el aeropuertollegada es igual al que buscamos lo agrega
                    subgraph.add_edge(node1, node2, weight=weight)  # añadimos la arista con el peso al grafo minimo

        geo = Nominatim(user_agent="MyApp", timeout=2) #se usa para el diccionario geopy


        plt.figure(figsize=(12, 8)) #grafica el mapa y tamaño
        my_map = Basemap(projection="robin", lon_0=0, lat_0=0)
        for i in path:
          lat = nodes[i][0].latitud
          lon = nodes[i][0].longitud
          x, y = my_map(lon, lat)
          pos[i]=(x,y)
        my_map.fillcontinents(color="grey")
        my_map.drawcountries()

        edge_labels = nx.get_edge_attributes(subgraph, 'weight')  #busca el peso para añadirlo al mapa
        nx.draw_networkx(subgraph, pos=pos, with_labels=True, node_size=50, node_color='blue')
        nx.draw_networkx_edge_labels(subgraph, pos=pos, edge_labels=edge_labels, font_size=10, bbox=dict(alpha=0))  #muestra los pesos de las aristas en el mapa

    plt.show()
    print("")
  elif menu == "5":
    geo = Nominatim(user_agent="MyApp",timeout=2)
    plt.figure(figsize=(12, 8))
    my_map = Basemap(projection="robin",lon_0=0, lat_0=0)
    my_map.fillcontinents(color="grey")
    my_map.drawcountries()
    for i in nodes:
            lat = nodes[i][0].latitud
            lon = nodes[i][0].longitud
            x, y = my_map(lon, lat)
            my_map.plot(x,y,marker='o',color='Red',markersize=3)
    plt.show()
    print("")

```
